```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the "Overall Cascade" column of the table
datasets = ['MapEval-Textual', 'MapEval-API', 'MapQA', 'MapEval-Visual']
overall_cascade_errors = [2.62, 3.55, 4.23, 4.36]

# Create the bar plot
plt.figure(figsize=(10, 6))
# Four different shades of blue
blue_shades = ['#81c2f5',  '#5ea8df',  '#3c8fc9', '#1f77b4']
bars = plt.bar(datasets, overall_cascade_errors, color=blue_shades)
# ['cornflowerblue', 'steelblue', 'darkcyan', 'teal']

# Add labels and title
plt.xlabel('Geospatial Benchmarks', fontsize=20)
plt.ylabel('Average Overall Cascade Error (%)', fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
# plt.title('Average Error (%) in MapAgent across 4 geospatial benchmarks', fontsize=14)

# Add the values on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval:.2f}', ha='center', va='bottom', fontsize=20)

# Set y-axis limits to match the provided image
plt.ylim(0, 5)

# Add grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('mapagent_error_analysis_zoom.pdf')
# Display the plot
plt.show()
```
